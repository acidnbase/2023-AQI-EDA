```python
# Replace non-numeric values with NaN
df['Ozone'] = pd.to_numeric(df['Ozone'], errors='coerce')
df['PM25'] = pd.to_numeric(df['PM25'], errors='coerce')
df['CO'] = pd.to_numeric(df['CO'], errors='coerce')
df['PM10'] = pd.to_numeric(df['PM10'], errors='coerce')
df['NO2'] = pd.to_numeric(df['NO2'], errors='coerce')

# Convert NaN values to 0
df.fillna(0, inplace=True)

# Convert columns to integer
df['Ozone'] = df['Ozone'].astype(int)
df['PM25'] = df['PM25'].astype(int)
df['CO'] = df['CO'].astype(int)
df['PM10'] = df['PM10'].astype(int)
df['NO2'] = df['NO2'].astype(int)

# Check the data types after conversion
print(df.dtypes)
```

## 1. What is the overall trend of air quality (measured by AQI) over time across different cities and states?

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Group data by 'City' and 'State'
grouped = df.groupby(['city', 'state'])

# Plotting the trend for each group
plt.figure(figsize=(12, 8))
for name, group in grouped:
    plt.plot(group['Date'], group['Overall AQI Value'], label=name)

plt.title('Overall Trend of Air Quality (AQI) Over Time')
plt.xlabel('Date')
plt.ylabel('Overall AQI Value')
plt.legend(title='city, state', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### The AQI Values across different States & Cities range closely to round 250 however, we can observe huge spike during month of October exceding values over as close to 700.

## 2.Which pollutant(s) contribute most to poor air quality in different cities?

```python
# Group data by 'City' and calculate the average AQI for each pollutant
avg_aqi_by_pollutant = df.groupby('city').agg({
    'Ozone': 'mean',
    'PM25': 'mean',
    'CO': 'mean',
    'PM10': 'mean',
    'NO2': 'mean'
})

# Calculate total average AQI for each city
avg_aqi_by_pollutant['Total AQI'] = avg_aqi_by_pollutant.mean(axis=1)

# Identify the pollutant(s) with the highest average AQI for each city
pollutant_contributions = avg_aqi_by_pollutant.idxmax(axis=1)

print("Pollutant(s) contributing most to poor air quality in different cities:")
print(pollutant_contributions)
```

### Analysizing different pollutatnts across cities we can observe a wide range of affectors across different Cities.

## 3. Are there any seasonal patterns in air quality across different cities and states?

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Extract month from 'Date' column and map it to seasons
df['Month'] = df['Date'].dt.month
df['Season'] = df['Month'].apply(get_season)

# Group data by 'City' and 'State' and 'Season', and calculate the average AQI for each season
seasonal_avg_aqi = df.groupby(['city', 'state', 'Season']).agg({'Overall AQI Value': 'mean'}).reset_index()

# Pivot the table for better visualization
pivot_seasonal_avg_aqi = seasonal_avg_aqi.pivot_table(index=['city', 'state'], columns='Season', values='Overall AQI Value')

# Display the seasonal average AQI for each city and state
print(pivot_seasonal_avg_aqi)
```

### Air Quality range different for different cities but, usually Winter & Fall tend to have low AQIs in some cities while major chunk of cities have higher AQI during Summer Season

## 4. What are the most common AQI categories observed across different cities and states?

```python
# Group data by 'City', 'State', and 'AQI category', and count occurrences
aqi_category_counts = df.groupby(['city', 'state', 'AQI category']).size().reset_index(name='Count')

# Find the most common AQI category for each city and state
most_common_aqi_category = aqi_category_counts.groupby(['city', 'state']).apply(lambda x: x.loc[x['Count'].idxmax()]).reset_index(drop=True)

# Display the most common AQI category for each city and state
print(most_common_aqi_category)
```

### While most cities tend to enjoy a Good quality AQI where else, a hand-full of Cities have moderate category of AQI

## 5. How does air quality differ between different monitoring sites within the same city?

```python

# Group data by 'City' and 'Name of site', and calculate the average AQI for each site
avg_aqi_by_site = df.groupby(['city', 'Site Name (of Overall AQI)']).agg({'Overall AQI Value': 'mean'}).reset_index()

# Display the average AQI for each monitoring site within the same city
print(avg_aqi_by_site)
```

### Air Quality in different parts of the same city varies from area to area. Some have an AQI Value above 60 whereas some tend to have as close as 40-45.
